```python
#---Loading the Dataset From a File

import pandas as pd
brainFile = 'brainsize.txt'
brainFrame = pd.read_csv(brainFile, delim_whitespace=True)

#---Verifying the Dataframe
brainFrame.head()
```

```python
#---Using the desribe() method

brainFrame.describe()
```

```python
#---Loading the Required Modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
#---Separate the Data

menDf = brainFrame[(brainFrame.Gender == 'Male')]
womenDf = brainFrame[(brainFrame.Gender == 'Female')]
```

```python
#---Plotting the graphs (For Men)
menMeanSmarts = menDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(menMeanSmarts, menDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
#---Plotting the graphs (For Women)
womenMeanSmarts = womenDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(womenMeanSmarts, menDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
#---Calculating correlation against brainFrame

brainFrame.corr(method='pearson')
```

```python
#---Calculating correlation against brainFrame for male-only dataframe

menDf.corr(method='pearson')
```

```python
#---Calculating correlation against brainFrame for female-only dataframe

womenDf.corr(method='pearson')
```

```python
#---Heatmap for female only dataset

import seaborn as sns
wcorr = womenDf.corr()
sns.heatmap(wcorr)
#plt.savefig('attribute_correlations.png', tight_layout=True)
```

```python
#---Heatmap for male only dataset

mcorr = menDf.corr()
sns.heatmap(mcorr)
#plt.savefig('attribute_correlations.png', tight_layout=True)
```
